# Sintaxis Básica del Lenguaje Go

El lenguaje Go, también conocido como Golang, es un lenguaje de programación moderno, creado por Google en 2007 y anunciado públicamente en 2009. Diseñado con el propósito de combinar la simplicidad, el rendimiento y la eficiencia, Go se ha convertido en una herramienta esencial para el desarrollo de software escalable, rápido y confiable.

Go se destaca por sus características únicas, tales como:

Simplicidad: Una sintaxis clara y minimalista que facilita el aprendizaje y la escritura de código.
Eficiencia: Concurrencia incorporada a través de goroutines y canales, que permite manejar múltiples tareas de manera eficiente.
Compilación Rápida: Un compilador altamente optimizado que genera binarios ejecutables ligeros.
Portabilidad: Compatibilidad en múltiples plataformas, ideal para el desarrollo de sistemas distribuidos y aplicaciones web modernas.
Esta documentación tiene como objetivo proporcionar una visión general y detallada del lenguaje Go, incluyendo Syntaxis básica del lenguaje Go (Variables, Bucles, Condiciones) .

## 1. Variables

Las variables en Go son fundamentales y un poco diferentes a otros lenguajes. Lo genial es que tienes dos formas principales de declararlas, dependiendo de cuánto detalle quieras incluir:
Puedes declararlas con `var` o hacerlas rápido con `:=`. 

- **Declaración explícita:** Aquí defines el tipo de dato de forma manual. Es útil cuando quieres asegurarte de que la variable tenga un tipo específico.
- **Declaración implícita:** En este caso, Go detecta automáticamente el tipo según el valor que le asignes. Es más rápido y ahorra algo de escritura, pero no siempre es lo ideal para todos los casos.

**Ejemplo:**

```python
package main

import "fmt"

func main() {
    // Declaración explícita: Aquí especificamos el tipo de dato.
    var nombre string = "Juan"
    var edad int = 25

    // Declaración implícita: Go se da cuenta de qué tipo es.
    ciudad := "Madrid"

    fmt.Println("Nombre:", nombre)
    fmt.Println("Edad:", edad)
    fmt.Println("Ciudad:", ciudad)
}


```

Notas:
- Usa `var` cuando quieras ser súper claro con el tipo de dato.
- Usa `:=` cuando quieras hacer algo rápido. Pero ojo, esto solo funciona dentro de funciones.

## 2. Condicionales

En el lenguaje Go, las estructuras de control condicionales permiten tomar decisiones basadas en el resultado de expresiones lógicas. Estas estructuras son esenciales para controlar el flujo del programa y definir comportamientos dinámicos. Go implementa las condiciones de manera sencilla, evitando complejidades innecesarias y promoviendo una sintaxis clara y directa.

**Estructuras Condicionales Básicas** 
1. Uso de `if`
La sentencia `if` evalúa una expresión booleana. Si la condición es verdadera, ejecuta el bloque correspondiente.
2. Uso de `if-else`
Si la condición evaluada en `if` no es verdadera, el bloque else será ejecutado.
3. Uso de `else if`
Se utiliza `else if` para evaluar múltiples condiciones en cascada.

**Operadores Lógicos**
Go soporta los siguientes operadores lógicos para combinar múltiples condiciones:

`&&` (AND lógico): Verdadero si ambas condiciones son verdaderas.
`||` (OR lógico): Verdadero si al menos una condición es verdadera.
`!` (NOT lógico): Invierte el valor lógico.

**Ejemplo:**

```python
package main

import "fmt"

func main() {
    edad := 25
    tieneLicencia := true
    tieneSuspensiones := false

    // Evaluamos las condiciones combinadas
    if edad >= 18 && tieneLicencia && !tieneSuspensiones {
        fmt.Println("Puedes conducir legalmente.")
    } else if edad >= 18 && (!tieneLicencia || tieneSuspensiones) {
        fmt.Println("Tienes la edad para conducir, pero necesitas resolver tus problemas con la licencia o suspensiones.")
    } else {
        fmt.Println("No puedes conducir porque no cumples con los requisitos mínimos.")
    }
}


```

Notas rápidas:
- **Las llaves son obligatorias.** a diferencia de en otros lenguajes de programación

## 3. Bucles
Los bucles son una herramienta fundamental en cualquier lenguaje de programación para repetir tareas de manera eficiente. En Go, el único tipo de bucle disponible es `for`, pero se puede usar de diferentes formas para cubrir escenarios comunes, como bucles tradicionales, iteraciones sobre colecciones y bucles infinitos.

### 3.1. Bucle tradicional

El formato más común de un bucle `for` incluye una inicialización, una condición y un incremento. Este tipo de bucle es similar al de otros lenguajes como C o Java.

```python
package main

import "fmt"

func main() {
    for i := 0; i < 5; i++ {
        fmt.Println("Iteración:", i)
    }
}
```

### 3.2. Bucle con condición

En Go, no existe un bucle `while`, pero se puede lograr el mismo efecto utilizando un `for` con solo una condición.

```python
package main

import "fmt"

func main() {
    x := 0
    for x < 5 {
        fmt.Println("Valor de x:", x)
        x++
    }
}
```

### 3.3. Bucle infinito

Un bucle infinito se logra utilizando un `for` sin parámetros. Este tipo de bucle solo termina si se usa explícitamente una instrucción como `break`.

Consejo: **Siempre piensa bien si necesitas un bucle infinito.** Puede ser útil, pero también causar problemas si no lo manejas bien.

```python
package main

import "fmt"

func main() {
    for {
        fmt.Println("Esto es un bucle infinito")
        break // Rompe el bucle cuando sea necesario.
    }
}
```

## Fortalezas de Go
### Simplicidad y claridad:

Go se destaca por su sintaxis limpia y su enfoque minimalista, lo que facilita a los desarrolladores aprenderlo rápidamente y escribir código fácil de mantener.
Las decisiones de diseño del lenguaje, como la eliminación de características complejas innecesarias (herencia, genéricos avanzados hasta Go 1.18), lo hacen intuitivo y legible.
Alto rendimiento:

Compilado directamente a código máquina, Go combina la velocidad de lenguajes como C y C++ con una mayor facilidad de uso.
Su recolector de basura altamente optimizado garantiza un equilibrio entre rendimiento y facilidad de administración de memoria.
Concurrencia eficiente:

El modelo de concurrencia de Go, basado en goroutines y canales, lo convierte en una opción ideal para aplicaciones concurrentes y paralelas, como servidores web y microservicios.
Este enfoque simplifica la escritura de programas que aprovechan al máximo los procesadores multinúcleo.
Ecosistema robusto:

Go cuenta con una biblioteca estándar rica y herramientas integradas como el gestor de dependencias (go mod), herramientas de prueba (go test) y un servidor HTTP integrado.
Su comunidad activa y su soporte constante por parte de Google lo convierten en un lenguaje en evolución constante.
Compilación estática y despliegue sencillo:

Los binarios generados por Go son auto-contenidos, lo que elimina dependencias externas y facilita la distribución en múltiples plataformas.
