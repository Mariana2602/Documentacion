# Documentación Básica de Rust

Guía para comprender los pilares fundamentales de Rust.

## 1- Variables y mutabilidad

Las variables son inmutables por defecto. Esto promueve la seguridad y el manejo de memoria eficiente.

```rust
// Declarar variables inmutables.
let x = 5;

// Declarar variables mutables.
let mut y = 10;

// Imprimir el valor.
println!("Valor inicial: {}", y);

y = 15;
println!("Valor modificado: {}", y);

// Declarar constantes inmutables de tipo obligatorio.
const SEGUNDOS_EN_MINUTO: u32 = 60;
```

## 2- Tipos de Datos

Rust cuenta con tipos de datos robustos y estáticos, divididos principalmente en escalares (un solo valor: enteros, flotantes, booleanos, caracteres) y compuestos (agrupaciones: tuplas, arrays). Se destacan por tener tamaños fijos en memoria y un alto rendimiento. 

```rust
// Entero con signo
let entero: i32 = -42;        
// Punto flotante
let flotante: f64 = 3.1416;   
// Verdadero/Falso
let booleano: bool = true;    
// Carácter Unicode
let caracter: char = 'z';     

// Compuestos (Tuplas y Arrays)
let tupla: (i32, f64, &str) = (500, 6.4, "Hola");
let arreglo: [i32; 3] = [10, 20, 30]; // Tamaño fijo

println!("El primer elemento del arreglo es: {}", arreglo[0]);
```

## 3- Control de Flujo

El control de flujo en Rust se basa en expresiones if/else, emparejamiento de patrones con match y bucles (loop, while, for). Las condiciones deben ser estrictamente booleanas. Destaca por permitir usar if y match para retornar valores directamente, y por la seguridad del sistema de tipos.

```rust
let numero = 7;

if numero < 5 {
    println!("Es pequeño");
} else {
    println!("Es grande");
}

println!("Contando:");
for i in 1..4 { 
    println!("{}", i);
}

let mut contador = 0;
let resultado = loop {
    contador += 1;
    if contador == 10 {
        break contador * 2; // El break puede devolver un valor
    }
};
println!("El resultado del loop es: {}", resultado);
```

## 4- Funciones

Las funciones en Rust se definen usando la palabra clave fn, siguen la convención snake_case, y requieren anotaciones de tipo para los parámetros. Son fundamentales para la reutilización de código, permiten retornar valores (usualmente la última expresión) y se ejecutan llamándolas por su nombre, siendo main el punto de entrada.

```rust
fn sumar(a: i32, b: i32) -> i32 {
    a + b  
}

let suma = sumar(5, 10);
println!("La suma es: {}", suma);
```

## 5- El Ownership (Propiedad)

El Ownership (propiedad) es el pilar fundamental de la gestión de memoria en Rust, permitiendo seguridad sin recolector de basura. Se basa en tres reglas clave: cada valor tiene un propietario, solo puede haber uno a la vez, y al salir de ámbito (scope), el valor se libera. Las reglas se verifican al compilar.

```rust
let s1 = String::from("hola");
let s2 = s1; // s1 ya no es válido.

// println!("{}", s1); // Esto causaría un error de compilación

// Para evitar esto, usamos Referencias
fn calcular_longitud(s: &String) -> usize {
    s.len()
}

let len = calcular_longitud(&s2); // Pasamos una referencia con '&'
println!("La longitud de '{}' es {}", s2, len); // s2 sigue siendo válido
```

## Conclusiones

Usar Rust concluye en un rendimiento excepcional similar a C/C++, garantizando seguridad de memoria y concurrencia sin recolector de basura gracias a su sistema de ownership. Es ideal para sistemas de alto rendimiento, WebAssembly y desarrollo web/embebido, aunque presenta una curva de aprendizaje pronunciada. Su compilador estricto previene errores en tiempo de ejecución, haciendo el código confiable y eficiente.
