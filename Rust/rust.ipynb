# Documentación Básica de Rust

Guía para comprender los pilares fundamentales de Rust.

## 1- Variables y mutabilidad

Las variables son inmutables por defecto. Esto promueve la seguridad y el manejo de memoria eficiente.

```python
// Declarar variables inmutables.
let x = 5;

// Declarar variables mutables.
let mut y = 10;

// Imprimir el valor.
println!("Valor inicial: {}", y);

y = 15;
println!("Valor modificado: {}", y);

// Declarar constantes inmutables de tipo obligatorio.
const SEGUNDOS_EN_MINUTO: u32 = 60;
```

## 2- Tipos de Datos

Rust cuenta con tipos de datos robustos y estáticos, divididos principalmente en escalares (un solo valor: enteros, flotantes, booleanos, caracteres) y compuestos (agrupaciones: tuplas, arrays). Se destacan por tener tamaños fijos en memoria y un alto rendimiento. 

```python
// Entero con signo
let entero: i32 = -42;        
// Punto flotante
let flotante: f64 = 3.1416;   
// Verdadero/Falso
let booleano: bool = true;    
// Carácter Unicode
let caracter: char = 'z';     

// Compuestos (Tuplas y Arrays)
let tupla: (i32, f64, &str) = (500, 6.4, "Hola");
let arreglo: [i32; 3] = [10, 20, 30]; // Tamaño fijo

println!("El primer elemento del arreglo es: {}", arreglo[0]);
```

## 3- Control de Flujo

El control de flujo en Rust se basa en expresiones if/else, emparejamiento de patrones con match y bucles (loop, while, for). Las condiciones deben ser estrictamente booleanas. Destaca por permitir usar if y match para retornar valores directamente, y por la seguridad del sistema de tipos.

```python
let numero = 7;

if numero < 5 {
    println!("Es pequeño");
} else {
    println!("Es grande");
}

println!("Contando:");
for i in 1..4 { 
    println!("{}", i);
}

let mut contador = 0;
let resultado = loop {
    contador += 1;
    if contador == 10 {
        break contador * 2; // El break puede devolver un valor
    }
};
println!("El resultado del loop es: {}", resultado);
```
